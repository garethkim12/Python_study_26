```python
#1. 숫자 4개를 입력받아 가장 큰수를 출력하는 코드를 작성하시오.
   #숫자 4개 입력 : 12 6 34 9
   #12, 6, 34, 9 중 가장 큰수 : 34
a, b, c, d = map(int, input("정수 4개 입력(예: 1 2 3 4): ").split()) # a,b,c,d 의 값을 각각 받아서 정수변환 후 4개의 값으로 분리해서 받음
max = a      #( 임의의 최댓값을 a로 지정하여 나머지 b,c,d 와 비교 검증하는 if로만 사용되는 조건문 생성
if max < b:  #(
   max = b   #(
if max < c:  #(
   max = c   #(
if max < d:  #(
   max = d   #(

print('최댓값:', max) #비교 검증 후 나온 최댓값 표기
     
```

```python
#2. 숫자 3개를 입력받아 작은수에서 큰수순으로 출력하는 코드를 작성하시오
  #숫자 3개를 입력 : 54 2 15
    #2  15  54
a, b, c = map(int, input('정수 3개를 입력하시오.:').split()) # 1번과 같이 정수 3개를 입력 받아 각각 분리 하여 받음
if a > b: # a, b 를 비교 하여 높은 숫자를 자리 바꿈
    a, b = b, a
if a > c: # a, c 를 비교 하여 높은 숫자를 자리 바꿈
    a, c = c, a
if b > c: # b, c 를 비교 하여 높은 숫자를 자리 바꿈
    b, c = c, b
print(c, b, a) #위 수식대로 진입 하였을 때 a<=b<=c 로 엮이므로 제일 높은 숫자 순서대로 출력 됨
```

```python

#3. 성적을 입력받아 학점을 출력하는 코드를 작성하시오
#   ---------------------------------
#        범위            학점
#   ----------------------------------
#    90 ~ 100            A
#    80 ~ 89             B
#    70 ~ 79             C
#    60 ~ 69             D
#    0 ~ 59              F
#   ----------------------------------
#
#   성적 입력 : 92
#   92점의 학점은 A입니다

#성적 입력
a = int(input('성적 입력:'))
# if 조건문을 이용한 각 점수 구간 별 학점 매기기
if 90 <= a <= 100:                        
    print(f"{a}점의 학점은 A 학점 입니다")  
if 80 <= a <= 89:                         
    print(f"{a}의 학점은 B 학점 입니다")     
if 70 <= a <= 79:
    print(f"{a}의 학점은 C 학점 입니다")
if 60 <= a <= 69:
    print(f"{a}의 학점은 D 학점 입니다")
if 0 <= a <= 59:
    print(f"{a}의 학점은 F 학점 입니다")    
```

```python
#5. 물건값을 입력하고, 지불한 돈의 액수를 입력했을 때 거스름돈을 출력한 후 
#1000원, 500원, 100원, 50원, 10원의 동전 개수를 출력하는 코드를 작성하시오.
#
#   물건값을 입력하시오 : 6370
#   지불한 돈의 액수를 입력하시오 : 10000
#
#   거스름돈 : 3630원
#   1000원 : 3개
#    500원 : 1개
#    100원 : 1개
#     50원 : 0개
#     10원 : 3개

# 물건값, 지불하는 금액 입력
a = int(input("물건값을 입력하시오 :"))
b = int(input("지불한 돈의 액수를 입력하시오 :"))
# 거스름 돈 출력
print(f'거스름돈 {b-a}원')
# 나눗셈 수식을 통한 1000원, 500원, 100원, 50원, 10원 단위의 거스름돈 출력
print(f'1000원 {(b-a) // 1000}개')
print(f'500원 {((b-a) % 1000) // 500}개')
print(f'100원 {(((b-a) % 1000) % 500) // 100}개')
print(f'50원 {((((b-a) % 1000) % 500) % 100) // 50}개')
print(f'10원 {(((((b-a) % 1000) % 500) % 100) % 50) // 10}개')
```

```python
#6. 5번 문제에서 지불한 돈의 액수가 물건값보다 작은 경우는 "XXX원이 부족합니다" 메세지 출력     
#    물건값을 입력하시오 : 6370
#    지불한 돈의 액수를 입력하시오 : 1000
#
#    5370원이 부족합니다

# 물건값, 지불하는 금액 입력
a = int(input("물건값을 입력하시오 :"))
b = int(input("지불한 돈의 액수를 입력하시오 :"))
#if 조건문을 통한 지불 금액 부족 시의 부족 금액 출력
if b-a < 0:
    print(f'{a-b}원 부족합니다')
    
# else 조건문을 이용하여 5번 문제와 동일한 값 출력
else:
    print(f'거스름돈 {b-a}원')
    print(f'1000원 {(b-a) // 1000}개')
    print(f'500원 {((b-a) % 1000) // 500}개')
    print(f'100원 {(((b-a) % 1000) % 500) // 100}개')
    print(f'50원 {((((b-a) % 1000) % 500) % 100) // 50}개')
    print(f'10원 {(((((b-a) % 1000) % 500) % 100) % 50) // 10}개')

```
