```python
# 1. 2 ~ 100사이의 소수를 출력하는 코드를 작성 (단, 한라인에 5개씩 출력)

#     2   3   5   7   11
#     13  17  19  23
#     ...
#     97


count = 0  # 출력한 소수 개수

for n in range(2, 101):          # 2~100
    c = True              # 일단 소수라고 가정

    for i in range(2, n):        # 2부터 n-1까지 나눠보기
        if n % i == 0:           # 나누어떨어지면 소수 아님
            c = False
            break

    if c:                 # 소수면 출력
        print(f"{n:3}", end=" ")
        count += 1

        if count % 5 == 0:       # 5개 출력마다 줄바꿈
            print()

# 마지막 줄이 5개로 딱 안 끝나면 줄바꿈 한번 해주기
if count % 5 != 0:
    print()
```
