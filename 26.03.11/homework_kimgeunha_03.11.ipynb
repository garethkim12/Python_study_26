```python
# 1. 학생 5명의 성적을 입력받아 다음과 같이 출력하는 코드를 작성

#     1's 학생의 성적 입력 : 80
#     80점의 학점은 B다

#     2's 학생의 성적 입력 : 1000
#     ERROR

#     3's 학생의 성적 입력 : 91
#     91점의 학점은 A다

#     4's 학생의 성적 입력 : 1000
#     ERROR

#     5's 학생의 성적 입력 : 69
#     69점의 학점은 D다

#     5명의 학생 입력 중 2번 에러발생
#     총점 : 80+91+69점
#     평균 : .xx점
```

```python
total = 0        # 정상 점수 총합
error = 0        # 에러 횟수

for i in range(1, 6):
    score = int(input(f"{i}'s 학생의 성적 입력 : "))

    if score < 0 or score > 100:
        print("ERROR")
        error += 1
    else:
        # 학점 계산
        if score >= 90:
            grade = "A"
        elif score >= 80:
            grade = "B"
        elif score >= 70:
            grade = "C"
        elif score >= 60:
            grade = "D"
        else:
            grade = "F"

        print(f"{score}점의 학점은 {grade}다")
        total += score
print(f"5명의 학생 입력 중 {error}번 에러발생")
print(f"총점 : {total}점")
print(f"평균 : {(total/5):.2f}점") #print("평균 : %.2f점" % (total/5))
```

```python
while True:
    n = int(input("2 - 100 사이의 정수 입력 : "))
    if 2 <= n <= 100:
        break
    else:
        print("범위를 벗어났습니다. 다시 입력하세요.")

# 입력된 정수의 약수 출력
print(f"정수 {n}의 약수 출력")

for i in range(1, n + 1):
    if n % i == 0:
        print(i, end=" ")
```

```python
for i in range(1, 4):
    n = int(input(f"{i}'s 정수 : "))

    if n < 2:
        print(f"{n}는 소수가 아닙니다")
        continue

    for j in range(2, n):
        if n % j == 0:
            print(f"{n}는 소수가 아닙니다")
            break
    else:
        print(f"{n}는 소수입니다")
```
